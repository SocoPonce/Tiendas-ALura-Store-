Visualización



**Ingresos promedio de cada tienda**

```python
#Ingresos.plot(kind='barv',figsize=(15,5),color='blue')
Ingresos = [tienda_1, tienda_2, tienda_3, tienda_4]

#df = pd.DataFrame(Ingresos, columns=['Ingresos'])

df = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingresos': Ingresos
})
df.plot(kind='barh', x='Tienda', y='Ingresos', figsize=(10,3), color='blue')
```

```python
df_tienda1_categoria_producto = tienda.groupby('Categoría del Producto')['Producto'].count()

df_tienda1_categoria_producto.plot(kind='bar', figsize=(10,3), color='pink')

plt.ylabel("Cantidad de Productos Vendidos")
plt.xlabel("Categoría del Producto")

plt.show()
```

Grafico del promedio de calificación de cada tienda.

```python
fig, ax1 = plt.subplots(figsize=(8,5))

ax1.bar(df['Tienda'], df['Calificación'], color='lightblue')
ax1.set_ylabel("Calificación")

ax2 = ax1.twinx()
ax2.plot(df['Tienda'], df['Calificación'], marker='o')
ax2.set_ylabel("Línea de Tendencia")

plt.title("Column + Line Chart")
plt.show()
```
